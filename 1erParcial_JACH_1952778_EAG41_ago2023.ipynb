```python
datos = pd.read_csv('https://raw.githubusercontent.com/jimmyzac/Estadistica-Aplicada-FCFM-UANL/main/bases_datos/poblacion_Asgard.csv')
datos
```

```python
# 2.a Ordena la base alfabéticamente por estado y por municipio
datos = datos.sort_values(['Estado', 'Municipio'], ascending=True)
datos
```

```python
# 2.b Cuantos estados hay, cuantos municipios hay en cada estado y cuantos municipios hay en total.
estados = datos.value_counts('Estado')
estados.count()
print('En total hay', estados.count(), 'estados')
```

```python
# 2.b Cuantos estados hay, cuantos municipios hay en cada estado y cuantos municipios hay en total
muni_estados= datos.groupby('Estado').count()
muni_estados[['Municipio']]
```

```python
# 2.b Cuantos estados hay, cuantos municipios hay en cada estado y cuantos municipios hay en total.
municipios = datos['Municipio'].count()
municipios
print('En total hay', municipios, 'municipios')
```

```python
# 2.c Cuanta población tiene el país en total y cuanta población tiene cada estado
poblacion = datos['Población'].sum()
poblacion
print('En total hay', poblacion, 'personas')
```

```python
# 2.c Cuanta población tiene el país en total y cuanta población tiene cada estado
estad_pobla = datos.groupby('Estado').sum()
estad_pobla['Población']
```

```python
# 2.d Calcula la mediana de la población de los municipios, y clasifícalos, aquellos por arriba de la mediana deben llamarse “muy poblados” y aquellos debajo de la mediana “poco poblados"
poblacion_muni= datos.groupby('Municipio').sum()
poblacion_muni['Población']
mediana = poblacion_muni['Población'].median()
poblacion_muni['Mediana']= np.where(poblacion_muni['Población']>mediana,'Muy Poblados','Poco Poblados')
poblacion_muni[['Población','Mediana']]
```

```python
# 2.e Encuentra los 5 municipios mas poblados y los cinco municipios menos poblados del país
poblacion_muni = poblacion_muni.sort_values('Población', ascending=False)
poblacion_muni[['Población', 'Mediana']].head()
```

```python
# 2.e Encuentra los 5 municipios mas poblados y los cinco municipios menos poblados del país
poblacion_muni[['Población', 'Mediana']].tail()
```

```python
# 2.f El municipio más poblado de cada estado
datos.groupby('Estado')[['Municipio', 'Población']].max()
```

```python
# 3.a Grafico de los 15 municipios más poblados de Asgard
graf = datos.groupby('Municipio').max('Población').sort_values('Población',ascending=False).head(15)
graf
graf['Población'].plot.bar()
plt.title('Municipios mas poblados de Asgard')
plt.xlabel('Municipio')
plt.ylabel('Poblacion')
plt.ticklabel_format(style='plain',axis='y')
plt.show()
```

```python
# 3.b Grafico de cuantas peliculas hay de cada género
generos= datos.groupby('category').count()
generos['title']
graf2= generos.sort_values('title', ascending=False)
graf2['title'].plot.bar()
plt.title('Peliculas por Genero')
plt.xlabel('Categoria')
plt.ylabel('Num de Peliculas')
plt.show()
```
