```python
## 1-Librerias
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# LIBRERIA ESPECIALIZADA
import statsmodels.formula.api as smf
```

```python
datos = pd.read_excel('modelolineal.xlsx')
datos = datos[['X','Y']]
```

```python
datos
```

estimar $Y = \beta_0 + \beta_1 X$

```python
## Especificamos el modelo
reg = smf.ols(formula='Y~X', data=datos)
results = reg.fit()
print(results.summary())
```

Para evaluar un modelo generalmente se usa $R^2$. Pero no es la unica medida utilizada, tambien se suele usar con bastante frecuencia el MAD, MAPE y MSD.

MAD: Desviacion absoluta media

MSD: Desviacion cuadratica media

MAPE: Error porcentual absoluto medio

MAD = Evalua y compara modelos

```python
## Definir el conjunto de prueba y el conjunto de evaluacion
## Generar x2
datos['X2'] = datos['X']**2
##
train_set = datos[:20]
test_set = datos[20:]
```

```python
train_set
```

## Estimar el primer modelo
estimar $Y = \beta_0 + \beta_1 X$

```python
## Definir el primer modelo
reg_m1 = smf.ols(formula='Y~X', data=train_set)
## estimamos
results_m1 = reg_m1.fit()
```

```python
b0 = results_m1.params[0]
b1 = results_m1.params[1]
```

```python
## Pronostico el conjunto de evaluacion
test_set['Y_hat'] = b0+b1*test_set['X']
```

```python
test_set
```

```python
# Estimar
MAD = (test_set['Y']-test_set['Y_hat']).abs().mean()
```

```python
print(MAD)
```

```python
## Estimar modelo 2
reg_m2 = smf.ols(formula='Y~X+X2', data=train_set)
## estimamos
results_m2 = reg_m2.fit()
##
b0 = results_m2.params[0]
b1 = results_m2.params[1]
b2 = results_m2.params[2]
## Pronostico el conjunto de evaluacion
test_set['Y_hat2'] = b0+b1*test_set['X']+b2*test_set['X2']
# Estimar
MAD2 = (test_set['Y']-test_set['Y_hat2']).abs().mean()

```

```python
print(MAD.__round__(2), MAD2.__round__(2))
```

El mejor modelo segun MAD y utilizando validacion cruzada es el modelo $Y = \beta_0 + \beta_1 X$
