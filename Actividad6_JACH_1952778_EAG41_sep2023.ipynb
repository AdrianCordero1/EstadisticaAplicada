# *3. Depurar la base de datos y estadísticas descriptivas*

```python
## 3.a) Verificar que no haya missing values, en caso de haberlos decidir imputar o eliminar
datos.isnull().sum()
datos = datos.dropna()
datos.isnull().sum()
```

```python
## 3.b) Verificar que no haya valores duplicados con dataframe.duplicated().sum()en caso de haber valores duplicados eliminarlos con dataframe=dataframe.drop_duplicates(keep='first').reset_index(drop=True)
datos.duplicated().sum()
datos = datos.drop_duplicates(keep='first').reset_index(drop=True)
datos.duplicated().sum()
```

```python
## 3.c) Verificar  que  todas  las  variables  sean  numéricas,  si  no  son  numéricas hacerlas   numéricas,   en   el   caso   de   las   categóricas   convertirlas   en dummies y conservar n-1 dummies
datos.dtypes
datos['satisfaccion'] = datos['satisfaccion'].astype('int64')
datos['evaluacion'] = datos['evaluacion'].astype('int64')
datos.dtypes
```

```python
dummies2 = pd.get_dummies(datos['salario'], prefix='sal').astype(int)
dummies2
```

```python
dummies2 = dummies2[['sal_bajo','sal_medio']]
datos = pd.concat([datos, dummies2], axis=1)
datos = datos.drop('salario', axis=1)
datos.dtypes
datos.isnull().sum()
```

# *6. Pruebe si el modelo de Cox cumple el supuesto de riesgos proporcionales, de no hacerlo estime el modelo AFT adecuado*

```python
cph1.check_assumptions(datos, p_value_threshold=0.05)
## H0: se viola el supuesto
## H1: se cumple
```

Como no se cumple el supuesto debemos utilizar un modelo alternativo, El modelo de riesgo acelerado AFT.

Primero debemos identificar si es exponencial, weibull o lognormal

```python
mexpo = ExponentialFitter().fit(datos['semestre'],datos['renunciar'])
mweibull = WeibullFitter().fit(datos['semestre'],datos['renunciar'])
mlogn = LogNormalFitter().fit(datos['semestre'],datos['renunciar'])
mexpo.AIC_, mweibull.AIC_, mlogn.AIC_
```

```python
## vamos a elegir el modelo LogNormal
from lifelines import LogNormalAFTFitter
lognAFT = LogNormalAFTFitter().fit(datos,'semestre','renunciar')
lognAFT.print_summary()
```
